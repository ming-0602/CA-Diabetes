Diabetes Dataset

The dataset is collected from the Iraqi society. The data consist of No. of Patient, Sugar Level Blood, Age, Gender, Creatinine ratio(Cr), Body Mass Index (BMI), Urea, Cholesterol (Chol), Fasting lipid profile, including total, LDL, VLDL, Triglycerides(TG) and HDL Cholesterol , HBA1C and Class of diabetes or non diabetes patient. The Objective of this study is to find the most potential characteristic of the diabeties. As World Health Organization show there are 422 million people wordwide which have diabeties and 1.5 million death are contributed by diabetes each year. As the study, it can show the effect of the characteristic of diabeties and improve the accurate approach to diagnosis the disease. 


What are the characteristic of diabeties?

Diabeties have multiple characteristic that can be determine such as Gender, Age , Urea, Cholesterol, etc. To determine these factor, it able to find the most obvious characteristic to help improve the diagnosis of diabeties.

Support Vector Machine

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values

x_train, x_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0) 


st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 


# "Support vector classifier"  
classifier = SVC(kernel='linear', random_state=0)  
classifier.fit(x_train, y_train)  

y_pred= classifier.predict(x_test) 

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)


plt.figure(figsize=(8, 6))

plt.scatter(x_test[y_test == 0, 0], x_test[y_test == 0, 1], color='blue', label='Actual Diabetise Negative: 0')

plt.scatter(x_test[y_test == 1, 0], x_test[y_test == 1, 1], color='orange', label='Actual Diabetise Positve: 1')

plt.scatter(x_test[y_pred == 0, 0], x_test[y_pred == 0, 1], color='green', marker='x', s=100, label='Predicted Diabetise Negative: 0')

plt.scatter(x_test[y_pred == 1, 0], x_test[y_pred == 1, 1], color='red', marker='x', s=100, label='Predicted Diabetise Positve: 1')

plt.title('Actual vs. Predicted')
plt.legend()
plt.show()
```

As the result of confusion matrix shown above is [[21 , 1] [4 , 244]]. 21 is the true positive data and the model have incorrectly predict 1 data only for the true positive data. As for the true negative data it have 224 true negative data and the model have only incorrectly predict 4 data only. This have shown the accuracy for the SVC model. 

As the accuracy_sorce shown it have 0.98 accuracy. The Maximum of accuracy score is 1 and the result shown above is 0.98. This have shown the data have a great accuracy. 

Decision Tree

```python
X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values

x_train, x_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0) 

decisiontree = DecisionTreeClassifier(random_state=42)

decisiontree.fit(x_train, y_train)

predictions = decisiontree.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print(accuracy)

print(classification_report(y_test, predictions))

plt.figure(figsize=(12, 8))
tree.plot_tree(decisiontree, feature_names=df.iloc[:, 3:-1].columns, class_names=['No Diabetes', 'Diabetes'], filled=True, rounded=True)
plt.show()
```

On the above it for the decision tree it have shown that it have a 0.996 out of 1. Which is a great accuracy in decision tree. Random Forest

```python
X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values

x_train, x_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0) 

randomtree = RandomForestClassifier(n_estimators=100, random_state=42)
randomtree.fit(x_train, y_train)

predictions = randomtree.predict(x_test)

accuracy = accuracy_score(y_test , predictions)
print(accuracy)

print(classification_report(y_test, predictions))
```

Random Forest is a ensemble learning method as well. As you can see it have 0.996 accuracy as well as Decision Tree. The prediction have show that the data provided is as accurate as it is.
