```python
df.rename(columns={"ID": "ID" ,"No_Pation": "No_Patient", "Gender" : "Gender (M/F)", "AGE" : "Age", "Urea" : "Urea (mmol/l)", "Cr" : "Creatinine ratio", "HbA1c" : "HbA1c (mmol/L)", "Chol" : "Cholesterol (mmol/l)", "TG" : "Triglycerides (mmol/l)", "HDL" : "High-density lipoprotein (mmol/l)", "LDL": "Low Desity Lipoprotein (mmol/l)", "VLDL" : "Very Low Desity Liporpotein (mmol/l)", "BMI" : "Body Mass Index (kg/m²)", "CLASS" : "Diabeties Positive"} , inplace=True)

df['Diabeties Positive'].replace('Y ', 'Y', inplace=True)
df['Diabeties Positive'].replace('N ', 'N', inplace=True)

df
```

Support Vector Machine

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix, accuracy_score


df['Diabeties Positive'].replace('N', 0, inplace=True)
df['Diabeties Positive'].replace('Y', 1, inplace=True)
df['Diabeties Positive'].replace('P', 1, inplace=True)
# df['Diabeties Positive'].replace('Y ', 1, inplace=True)
# df['Diabeties Positive'].replace('Y', 1)


X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values

x_train, x_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0) 


st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 


# "Support vector classifier"  
classifier = SVC(kernel='linear', random_state=0)  
classifier.fit(x_train, y_train)  

y_pred= classifier.predict(x_test) 

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)

```

As the result of confusion matrix shown above is [[21 , 1] [4 , 244]]. 21 is the true positive data and the model have incorrectly predict 1 data only for the true positive data. As for the true negative data it have 224 true negative data and the model have only incorrectly predict 4 data only. This have shown the accuracy for the SVC model. 

As the accuracy_sorce shown it have 0.98 accuracy. The Maximum of accuracy score is 1 and the result shown above is 0.98. This have shown the data have a great accuracy.
